# Création tables par thématique pour les BBA

```python
data = pd.read_excel('PB_2019_BBA.xlsx', sep = ';')
data
```

```python
# Thématiques et positions des colonnes pour chaque thématique
thematiques = [x for x in list(data) if not x.startswith('Unnamed')]
positions = dict()
for x in thematiques :
    positions[x] = data.columns.get_loc(x)
positions
```

```python
# liste avec les tables bachelor
bachelor_tables = list()
for i in range(len(positions)) :
    if i < len(positions) - 1 :
        bachelor_tables.append(data.iloc[:,positions.get(thematiques[i]) : positions.get(thematiques[i+1])])
    else :
        bachelor_tables.append(data.iloc[:, positions.get(thematiques[i]) : ])
```

```python
for i in range(len(bachelor_tables)) :
    bachelor_tables[i].reset_index(inplace = True)
    bachelor_tables[i].columns = bachelor_tables[i].iloc[0].tolist()
    bachelor_tables[i] = bachelor_tables[i] \
                                .loc[bachelor_tables[i]['identifiant école'].str.contains('BBA')] \
                                .reset_index(drop=True)
```

```python
for name, table in zip(thematiques, bachelor_tables) :
    table.to_csv("data_bba_2019/{}.csv".format(name), sep = ';')
```
