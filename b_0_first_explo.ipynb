```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('PB_2018_BBA.csv', sep = ';', encoding = "ISO-8859-1")
#data = data.loc[data['Nom du programme'].str.contains('BBA')].reset_index(drop = True)
```

```python
data
```

```python
thematiques = [x for x in list(data) if not x.startswith('Unnamed')]
positions = dict()
for x in thematiques :
    positions[x] = data.columns.get_loc(x)
positions
```

```python
bachelor_tables = list()
for i in range(len(positions)) :
    if i < len(positions) - 1 :
        bachelor_tables.append(data.iloc[:,positions.get(thematiques[i]) : positions.get(thematiques[i+1])])
    else :
        bachelor_tables.append(data.iloc[:, positions.get(thematiques[i]) : ])
```

```python
for i in range(len(bachelor_tables)) :
    bachelor_tables[i].reset_index(drop = True, inplace = True)
    bachelor_tables[i].columns = bachelor_tables[i].iloc[0].tolist()
```
