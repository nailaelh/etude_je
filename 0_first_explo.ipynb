# Bachelor

```python
bachelor = data.loc[~data.index.isin(["PB_M", "PP"])]
#bachelor = bachelor.reset_index()
bachelor = bachelor.reset_index(drop = True).set_index('level_1')
bachelor.index.name = None
```

```python
thematiques = [x for x in list(bachelor) if not x.startswith('Unnamed')]
positions = dict()
for x in thematiques :
    positions[x] = bachelor.columns.get_loc(x)
positions
```

```python
bachelor_tables = list()
for i in range(len(positions)) :
    if i < len(positions) - 1 :
        bachelor_tables.append(bachelor.iloc[:,positions.get(thematiques[i]) : positions.get(thematiques[i+1])])
    else :
        bachelor_tables.append(bachelor.iloc[:, positions.get(thematiques[i]) : ])
```

```python
for i in range(len(bachelor_tables)) :
    bachelor_tables[i].reset_index(inplace = True)
    bachelor_tables[i].columns = bachelor_tables[i].iloc[2].tolist()
```
