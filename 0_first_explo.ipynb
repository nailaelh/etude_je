```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("data_2018.xlsx", sheet_name="questionnaire")
data
```

```python
data = data.reset_index(level=1)
```

```python
bachelor = data.loc[~data.index.isin(["PB_M", "PP"])]
bachelor = bachelor.reset_index().set_index('level_1')
bachelor.index.name = None
bachelor = bachelor.drop(columns = 'index')
```

```python
thematiques = [x for x in list(bachelor) if not x.startswith('Unnamed')]
positions = dict()
for x in thematiques :
    positions[x] = bachelor.columns.get_loc(x)
```

```python
for position in positions :
    print(position)
```
